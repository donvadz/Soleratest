```python
if __name__ == '__main__':
    records = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        records.append([name,score])
    score = []
    for name in range(len(records)):
        score.append(records[name][1])
    score = sorted(score)
    smallest = score.pop(0)
    score1 = []
    for scored in range(len(score)):
        if score[scored] != smallest:
            score1.append(score[scored])
    second_minimum=score1[0]
    second_minimum_name = []
    for name in range(len(records)):
        if records[name][1] == second_minimum:
            second_minimum_name.append(records[name][0])
    second_minimum_name = sorted(second_minimum_name)
    for name in second_minimum_name:
        print(name)
            
```
